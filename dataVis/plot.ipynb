```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set()
```

```python
# import data generated by our go code
df = pd.read_csv("../resultData/results.csv")
df.head()
```

```python
# log plot of time complexity of 
# each sorting function we tested

sns.lineplot(
    x=df["N"],
    y=df["Wall Time"],
    hue=df["Function"],
    data=df,
)

plt.xscale('log')
# plt.yscale('log')
plt.title('Go Sorting Algorithm Performance')
plt.xlabel('Array Size (N)')
plt.ylabel('Wall Time (seconds)')
plt.legend(title='Sorting Algorithm')
plt.show()
```
