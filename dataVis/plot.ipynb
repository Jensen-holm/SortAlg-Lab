```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set()
```

```python
# import data generated by our go code
df = pd.read_csv("../resultData/results.csv")
print(df)
```

```python
# log plot of time complexity of 
# each sorting function we tested

sns.lineplot(
    x=df["N"],
    y=df["Wall Time"],
    hue=df["Function"],
    data=df,
)

plt.xscale('log', base=2)
# plt.yscale('log', base=2)
plt.title('Sorting Algorithm Performance')
plt.xlabel('Array Size (N)')
plt.ylabel('Wall Time (seconds)')
plt.legend(title='Sorting Algorithm')
plt.show()
```

### Report
The time complexity plot above makes sense given what we learned about each algorithm in class. Bubble sort has a time complexity of O(n^2), and in this plot you can clearly see that the bubble line (green) appears to be increasing with a relationship of n^2. For the other algorithms that have a logarithmic relationsip, the time to complete each algorithm is still increasing, but by a lot less than bubble sort. The bubble sort line is kind of skewing this plot, making it hard to see the increase in computation time for the other algorithms.
